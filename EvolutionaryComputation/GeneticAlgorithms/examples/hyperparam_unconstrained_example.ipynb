# Hyper Parameter Unconstraiend Optimization Examples

```python
# load in necessary libraries
import EvolutionaryComputation.OptimizationFunctions as opt_fun
import EvolutionaryComputation.GeneticAlgorithms as ga
```

## Sphere Function

```python
# define optimization function
fitness_function = opt_fun.sphere_function  # minimum at x=0 vector

# number of dimensions
d = 200

# define bounds
lower_bound = [-5.12]*d
upper_bound = [5.12]*d

# define number of solutions per generation
gen_size = 200
```

```python
# define the algorithm
algorithm = ga.HyperParamUnconstrainedProblem(fitness_function=fitness_function, 
                                          upper_bound=upper_bound, lower_bound=lower_bound,
                                          gen_size=gen_size)
```

### Parameter Set 1

```python
algorithm.evolve(p_cross=0.75, p_mutate=0.75, mutate_bound=0.1, elitism=0.2, p_method='static', mut_method='static',
                sel_method='roulette', tourn_size=0.2, max_iter=100, par_battle=False, find_max=False, info=False)
algorithm.plot(starting_gen=20)
algorithm.best_individual
```
