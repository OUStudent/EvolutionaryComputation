# Constrained Optimization Example

## Pressure Vessel Design Problem

The objective is to minimize the cost of material by optimizing the design of a pressure vessel. Here is a basic diagram of the variables available to optimize:



In order to satisify physical constraints, the design variables must observe the following cosntriants:



```python
# load in necessary libraries
import EvolutionaryComputation.OptimizationFunctions as opt_fun
import EvolutionaryComputation.GeneticAlgorithms as ga
```

```python
# define bounds for Pressure Vessel
upper_bound = [6.1875, 6.1875, 200, 200]
lower_bound = [0.0625, 0.0625, 10, 10]

# define fitness function
fitness_function = opt_fun.pressure_vessel

# define constraints
constraints = opt_fun.constraints_pressure_vessel

# define number of solutions per generation
gen_size = 200
```

```python
algorithm = ga.ConstrainedProblem(fitness_function=fitness_function, constraints=constraints, upper_bound=upper_bound,
                           lower_bound=lower_bound, gen_size=gen_size)
```

```python
algorithm .evolve(find_max=False, max_iter=400)
```

```python
algorithm.plot()
algorithm.plot(plot_sigmas=True)
```

Note that each strategy paramter refers to the variation of that respective variable. For example, the second plot below, Mean Strategy Paramter 1, showcases the mean variation of the population for the first variable. We can see that towards the end of evolution, generations $300$ to $400$, the mean variation of that particular variable increased from roughly $0.00250$ to 0.00350. This means that towards the end of evolution, the first variable was mutated by some $N(0, 1)$ value times $0.00350$; if N(0, 1) was $0.75$, then x1 was mutated by $0.75*0.00350=0.002625$. 

```python
algorithm.plot(starting_gen=250)
algorithm.plot(plot_sigmas=True, starting_gen=250)
```
